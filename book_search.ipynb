```python
import pandas as pd
df=pd.read_csv('books.csv')
df.head()
```

```python
#selecting relevant data
df = df[['title', 'authors', 'original_publication_year', 'language_code', 'average_rating']]
# Removing missing data
df.dropna(inplace=True)
# Reseting index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
def search_books(keyword, column='title'):
    """
    Search for books containing the keyword in the specified column.
    Returns top 10 matches.
    """
    # Filter rows where the column contains the keyword (case-insensitive)
    results = df[df[column].str.contains(keyword, case=False, na=False)]
    
    # Select and return relevant columns
    return results[['title', 'authors', 'original_publication_year', 'average_rating']].head(10)
def search_books(keyword, column='title'):
    '''
    search for books containing the keyword in the sepcified column.
    returns the top 10 matches.
    '''
    #filtering the rows where the column contains the keyword (case-insensitive)
    results=df[df[column].str.contains(keyword, case=False, na=False)]
    return results
```

```python
def filter_by_year(year):
    """
    Return top 10 books published in the specified year.
    """
    results = df[df['original_publication_year'] == year]
    return results[['title', 'authors', 'original_publication_year', 'average_rating']].head(10)
```

```python
def search_multiple(keyword):
    results_title = df[df['title'].str.contains(keyword, case=False, na=False)]
    results_authors = df[df['authors'].str.contains(keyword, case=False, na=False)]
    combined = pd.concat([results_title, results_authors]).drop_duplicates()
    return combined[['title', 'authors', 'original_publication_year', 'average_rating']].head(10)
```

```python
def search_and_sort(keyword, column='title'):
    """
    Search books by keyword in the specified column,
    then sort by average_rating descending.
    """
    results = df[df[column].str.contains(keyword, case=False, na=False)]
    results = results.sort_values(by='average_rating', ascending=False)
    return results[['title', 'authors', 'original_publication_year', 'average_rating']].head(10)
```

```python
def search_and_sort(keyword, column='title'):
    results = df[df[column].str.contains(keyword, case=False, na=False)]
    results = results.sort_values(by='average_rating', ascending=False)
    return results[['title', 'authors', 'original_publication_year', 'average_rating']].head(10)
```

```python
def user_search():
    keyword = input("Enter a keyword to search: ")
    column = input("Search by 'title' or 'authors'? ").strip().lower()
    if column not in ['title', 'authors']:
        print("Invalid column, defaulting to title.")
        column = 'title'
    results = search_and_sort(keyword, column)
    print(results)
    save = input("Save results to CSV? (yes/no): ").strip().lower()
    if save == 'yes':
        filename = input("Enter filename (e.g., results.csv): ")
        save_search_results(results, filename)
    else:
        print("Results not saved.")
def save_search_results(results, filename='search_results.csv'):
    """
    Save a DataFrame of search results to a CSV file.
    """
    results.to_csv(filename, index=False)
    print(f"Results saved to {filename}")
```

```python
user_search()
```

```python
import random

def book_recommendation_quiz():
    print("Welcome to the Book Recommendation Quiz!")
    
    era = input("Do you prefer classic (before 2000) or modern (2000 and after) books? Enter 'classic', 'modern', or 'any': ").strip().lower()
    high_rating = input("Do you want only highly rated books? (yes/no): ").strip().lower()
    keyword = input("Enter a keyword to filter by title or author (or press Enter to skip): ").strip().lower()
    
    # Start filtering the dataframe
    filtered = df.copy()
    
    # Filter by era
    if era == 'classic':
        filtered = filtered[filtered['original_publication_year'] < 2000]
    elif era == 'modern':
        filtered = filtered[filtered['original_publication_year'] >= 2000]
    # else 'any' - no filter on year
    
    # Filter by rating
    if high_rating == 'yes':
        filtered = filtered[filtered['average_rating'] >= 4.0]
    
    # Filter by keyword in title or authors
    if keyword:
        filtered = filtered[
            filtered['title'].str.lower().str.contains(keyword) | 
            filtered['authors'].str.lower().str.contains(keyword)
        ]
    
    # If less than 3 books found, adjust
    if len(filtered) == 0:
        print("Sorry, no books found matching your criteria.")
        return
    elif len(filtered) < 3:
        sample_size = len(filtered)
    else:
        sample_size = 3
    
    # Pick random books
    recommendations = filtered.sample(sample_size)
    
    print("\nHere are your book recommendations:")
    for idx, row in recommendations.iterrows():
        print(f"- {row['title']} by {row['authors']} ({row['original_publication_year']}), Rating: {row['average_rating']}")
```

```python
book_recommendation_quiz()
```
